```python
import os
import pandas as pd
from pandasql import sqldf

consolePath = f"../Isaiah/"
lookupPath = f"../Isaiah/LOOKUP FILES"

outputDir = f"../Output/"
os.makedirs(outputDir, exist_ok=True)

print(consolePath)
print(lookupPath)

columnNames = ['BVN','FirstName','MiddleName','LastName','Email','Gender','NIN','Phone1','Phone2']
lookupNames = ["CUST_NO","BVN"]
lookupDTypes = {0: str, 1: str}
```

```python
consoleDf = []
lookupDF = []

for consoleFiles in os.listdir(consolePath):
    if consoleFiles.startswith('BVN_CONSOLE_NEW') and consoleFiles.endswith('.csv'):
        consPath = os.path.join(consolePath, consoleFiles)
        consDf = pd.read_csv(consPath, skiprows=1, names=columnNames, on_bad_lines='skip')
        consoleDf.append(consDf)

for lookupFiles in os.listdir(lookupPath):
    if lookupFiles.startswith('BVN_CONSOLE_FILE') and consoleFiles.endswith('.csv'):
        lookPath = os.path.join(lookupPath, lookupFiles)
        lookDf = pd.read_csv(lookPath, skiprows=1, names=lookupNames, on_bad_lines='skip', dtype=lookupDTypes)
        lookupDF.append(lookDf)

cons = pd.concat(consoleDf, ignore_index=True)
look = pd.concat(lookupDF, ignore_index=True)
```

```python
display(cons.head(10))
display(look.head(10))
```

```python
consQL = """ select a.cust_no, a.bvn, b.FirstName, b.MiddleName,b.LastName, b.Email,
    b.Gender, b.NIN, b.Phone1,b.Phone2 from look a left join cons b on a.bvn = b.bvn"""
consqlDf = sqldf(consQL, locals())
display(consqlDf.head(10))
```

### Write to Output

```python
finalPath = os.path.join(outputDir, "finalFiles.csv")

consqlDf.to_csv(finalPath, index=False, mode='w')
print(f"Final file {finalPath} written to output path")
```
